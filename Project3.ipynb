```python
import requests
from google.colab import userdata

# KEYS (replace with yours)
SPEECH_KEY = userdata.get('speech')
SPEECH_REGION = "koreacentral"

VISION_KEY =  userdata.get('vision')
VISION_ENDPOINT = "https://koreacentral.api.cognitive.microsoft.com/"

OPENAI_KEY = userdata.get('openai')
OPENAI_ENDPOINT = "https://smart-classroom-ai-assi-resource.services.ai.azure.com/"
DEPLOYMENT = "gpt-4.1"


#  1. Speech → Text (simplified placeholder)
def speech_to_text(audio_file):
    # In exam, you can assume transcription done
    return "Explain this diagram"


# 2. Image → Description
def analyze_image(image_path):
    url = f"{VISION_ENDPOINT}computervision/imageanalysis:analyze?api-version=2024-02-01&features=caption"

    headers = {
        "Ocp-Apim-Subscription-Key": VISION_KEY,
        "Content-Type": "application/octet-stream"
    }

    with open(image_path, "rb") as f:
        response = requests.post(url, headers=headers, data=f)

    result = response.json()

    print(result)  # debug

    if "error" in result:
        return "Image analysis failed"

    if "captionResult" in result:
        return result["captionResult"]["text"]

    return "No description available"

# 3. OpenAI Response
def generate_answer(student_id, speech_text, image_desc):

    prompt = f"""
Student ID: {student_id}

Student asked: {speech_text}
Image description: {image_desc}

Explain clearly.
Always include Student ID in response.
"""

    url = f"{OPENAI_ENDPOINT}/openai/deployments/{DEPLOYMENT}/chat/completions?api-version=2024-02-15-preview"

    headers = {
        "Content-Type": "application/json",
        "api-key": OPENAI_KEY
    }

    data = {
        "messages": [
            {"role": "system", "content": "You are a smart classroom assistant."},
            {"role": "user", "content": prompt}
        ],
        "max_tokens": 600
    }

    response = requests.post(url, headers=headers, json=data)
    result = response.json()

    answer = result['choices'][0]['message']['content']

    return f"[Student ID: {student_id}]\n{answer}"


# MAIN PIPELINE
def smart_assistant(student_id, audio_file, image_path):
    speech_text = speech_to_text(audio_file)
    image_desc = analyze_image(image_path)

    return generate_answer(student_id, speech_text, image_desc)


# Example
print(smart_assistant("BTECH2026_333", "Sports.wav", "Computer_vision2.png"))
```
