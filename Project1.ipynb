```python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

from google.colab import userdata
api_key=userdata.get('Cap')


# --- 1. CONFIGURATION ---
# Found in Azure AI Foundry: Settings -> Project Connection / Azure OpenAI Keys
AZURE_ENDPOINT = "https://university-help-desk-pr-resource.openai.azure.com/"
API_KEY = api_key
API_VERSION = "2024-02-01"
MODEL_DEPLOYMENT_NAME = "gpt-4.1" # The name YOU gave your deployment

# Initialize the Azure Client
client = AzureOpenAI(
    azure_endpoint=AZURE_ENDPOINT,
    api_key=API_KEY,
    api_version=API_VERSION
)

# --- 2. SYSTEM PROMPT (The "Brain") ---
SYSTEM_INSTRUCTIONS = """
You are a University AI Helpdesk Assistant.
MANDATORY RULE: Every response must start with the Student ID in brackets.
Example: [Student ID: BTECH2026_101]
Only answer queries related to Admissions, Fees, Hostel, and Courses.
If the query is irrelevant or harmful, politely decline.
"""

def ask_helpdesk(student_id, user_query):
    try:
        # Create the chat completion
        response = client.chat.completions.create(
            model=MODEL_DEPLOYMENT_NAME,
            messages=[
                {"role": "system", "content": SYSTEM_INSTRUCTIONS},
                {"role": "user", "content": f"Student ID: {student_id}\nQuery: {user_query}"}
            ],
            temperature=0.3, # Keeps responses factual and consistent
            max_tokens=500
        )

        # Return the actual text answer
        return response.choices[0].message.content

    except Exception as e:
        return f"Error connecting to Azure: {str(e)}"

# --- 3. RUN THE TEST ---
if __name__ == "__main__":
    # Simulate your input JSON
    student_data = {
        "student_id": "BTECH2026_101",
        "query": "What is the Course Structure?"
    }

    print("Sending query to Azure AI...")
    print("-" * 30)

    result = ask_helpdesk(student_data["student_id"], student_data["query"])

    print(result)
```
